```python
tn=pd.read_csv('/kaggle/input/playground-series-s3e1/train.csv')
tt=pd.read_csv('/kaggle/input/playground-series-s3e1/test.csv')
tn.head()
```

```python
x=tn.drop(columns=['MedHouseVal','id','AveBedrms'])
y=tn['MedHouseVal']
```

```python
id=tt['id']
```

```python
tt=tt.drop(columns=['id','AveBedrms'])
```

```python
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor(iterations=1750, bagging_temperature=0.8)
model.fit(x,y)
y_pred=model.predict(tt)
```

```python
sub={"Id":id,"MedHouseVal":y_pred}
sub=pd.DataFrame(sub)
```

```python
sub.to_csv("submission.csv",index=False)
```
